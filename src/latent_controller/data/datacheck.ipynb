```python
import plotly_express as px
```

```python
def hex_to_rgb(hex):
    hex = str(hex).strip("#")
    return 'rgb' + str(tuple(int(hex[i:i+2], 16) for i in (0, 2, 4)))
```

```python
color_seq = [hex_to_rgb(x) for x in df['colour_n=0'].tolist()]
```

```python
fig = px.scatter(df,
                 x='x_emb_n=0',
                 y='y_emb_n=0',
                 hover_name='id',
                 color='id',
                 color_discrete_sequence=df['colour_n=0'].tolist())
```
